```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/crop_yield.csv')
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['soil_quality'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Soil Quality')
plt.ylabel('Frequency')
plt.title('Distribution of Soil Quality')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['rainfall'], bins=10, color='lightgreen', edgecolor='black')
plt.xlabel('Rainfall')
plt.ylabel('Frequency')
plt.title('Distribution of Rainfall')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='seed_type', data=df, palette='Set2')
plt.xlabel('Seed Type')
plt.ylabel('Count')
plt.title('Seed Type Distribution')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['soil_quality'], df['rainfall'], alpha=0.7, color='orange')
plt.xlabel('Soil Quality')
plt.ylabel('Rainfall')
plt.title('Soil Quality vs. Rainfall')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='yield_category', y='rainfall', data=df, palette='pastel')
plt.xlabel('Yield Category')
plt.ylabel('Rainfall')
plt.title('Rainfall by Yield Category')
plt.tight_layout()
plt.show()
```
